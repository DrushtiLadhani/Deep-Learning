```python
total_train=num_dcl_tr+num_dcp_tr+num_fcl_tr+num_fcp_tr
total_val=num_dcl_val+num_dcp_val+num_fcl_val+num_fcp_val
print('Total Training images = ',total_train)
print('Total validation images = ',total_val)
```

```python
batch_size = 200
epochs = 50
IMG_HIEGHT = 150
IMG_WIDHT = 150
```

```python
train_image_generator = ImageDataGenerator(rescale=1./255)
validation_image_generator = ImageDataGenerator(rescale=1./255)
```

```python
train_data_gen = train_image_generator.flow_from_directory(batch_size=batch_size,
                                                           directory='C:\\Users\\admin\\.keras\\datasets\\cotton_plant_disease\\train',
                                                           shuffle=True,
                                                           target_size=(IMG_HIEGHT,IMG_WIDHT),
                                                           class_mode='binary')
```

```python
train_data_gen.class_indices
```

```python
val_data_gen = validation_image_generator.flow_from_directory(batch_size=batch_size,
                                                              directory='C:\\Users\\admin\\.keras\\datasets\\cotton_plant_disease\\val',
                                                              target_size=(IMG_HIEGHT,IMG_WIDHT),
                                                              class_mode='binary')
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1,5,figsize=(20,20))
    axes = axes.flatten()
    for img,ax in zip(images_arr,axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
plotImages(sample_image[0:5])
```

```python
model = Sequential([
    Conv2D(16,3,padding='same',activation='relu',input_shape=(IMG_HIEGHT,IMG_WIDHT,3)), #150
    MaxPooling2D(), #75
    Conv2D(32,3,padding='same',activation='relu'), #75
    MaxPooling2D(),#37
    Conv2D(64,3,padding='same',activation='relu'),#37
    MaxPooling2D(),#18
    Flatten(),
    Dense(512,activation='relu'),
    Dense(4,activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(train_data_gen,steps_per_epoch=total_train//batch_size,epochs=epochs,
                             validation_data=val_data_gen,validation_steps=total_val//batch_size)
```

```python
history.history
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(epochs)
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.plot(epochs_range,acc,label='Training Accuracy')
plt.plot(epochs_range,val_acc,label='Val Accuracy')
plt.legend(loc=0)
plt.title('Accuracy')
plt.subplot(1,2,2)
plt.plot(epochs_range,loss,label='Training Loss')
plt.plot(epochs_range,val_loss,label='Val LOss')
plt.legend(loc=0)
plt.title('Loss')
plt.show()
```

```python
image_gen_val = ImageDataGenerator(rescale=1./255)
val_data_gen = image_gen_val.flow_from_directory(batch_size=batch_size,
                                                 directory='C:\\Users\\admin\\.keras\\datasets\\cotton_plant_disease\\val',
                                                 target_size=(IMG_HIEGHT,IMG_WIDHT),
                                                 class_mode='binary')
```

```python
model_new = Sequential([
    Conv2D(16,3,padding='same',activation='relu',input_shape=(IMG_HIEGHT,IMG_WIDHT,3)), #150
    MaxPooling2D(), #75
    Dropout(0.2),
    Conv2D(32,3,padding='same',activation='relu'), #75
    MaxPooling2D(),#37
    Dropout(0.2),
    Conv2D(64,3,padding='same',activation='relu'),#37
    MaxPooling2D(),#18
    Dropout(0.2),
    Flatten(),
    Dense(512,activation='relu'),
    Dropout(0.2),
    Dense(4,activation='softmax')
])
```

```python
model_new.summary()
```

```python
epochs= 500
```

```python
model_new.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model_new.fit(train_data_gen,steps_per_epoch=total_train//batch_size,
                                 epochs=epochs,
                                 validation_data=val_data_gen,
                                 validation_steps=total_val//batch_size)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(epochs)
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.plot(epochs_range,acc,label='Training Accuracy')
plt.plot(epochs_range,val_acc,label='Val Accuracy')
plt.legend(loc=0)
plt.title('Accuracy')
plt.subplot(1,2,2)
plt.plot(epochs_range,loss,label='Training Loss')
plt.plot(epochs_range,val_loss,label='Val LOss')
plt.legend(loc=0)
plt.title('Loss')
plt.show()
```
